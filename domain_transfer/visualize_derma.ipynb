```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import os

current_dir = os.path.dirname(os.path.abspath('__file__'))
format_file = os.path.join(current_dir, '.matplotlibrc')
with mpl.rc_context(fname=format_file):
    # Data setup
    expertise_levels = ["AI Assistant", "Medical Assistant", "General Physician", "Dermatologist"]
    baseline_gpt4 = [30, 35, 45, 55]

    # X coordinates for points
    x = np.arange(len(expertise_levels))

    fig, ax1 = plt.subplots(figsize=(4, 3))
    # Plot persona accuracy line
    line1, = ax1.plot(x, baseline_gpt4, marker='o', linestyle='-', color='mediumseagreen', 
                      label="Zero-shot (ZS)", markersize=7, linewidth=2)

    # Add a dark green blob for CoT accuracy (all personas, mean value)
    ax1.scatter([3], [70], color='darkgreen', s=50, label="Chain-of-thought (CoT)", zorder=5)

    # X-ticks and labels
    ax1.set_xticks(x)
    ax1.set_xticklabels(expertise_levels, rotation=15)

    ax1.set_xlabel("Persona")
    ax1.set_ylabel("Accuracy (%)")
    ax1.set_ylim(0, 80)
    ax1.grid(False)
    ax1.legend(loc="lower right")

    fig.tight_layout()
    plt.show()
```
